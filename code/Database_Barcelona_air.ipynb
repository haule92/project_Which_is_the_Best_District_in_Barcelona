```python
import pandas as pd
import numpy as np
air_table = pd.read_csv("air_quality_Nov2017.csv")
air_table_clean = air_table.copy()
air_table_clean = air_table_clean[["Air Quality", "Station","Longitude"]]
air_table_clean = air_table_clean.groupby(["Station", "Air Quality"]).count()[["Longitude"]]
air_table_clean = air_table_clean.reset_index()
air_table_clean
```
